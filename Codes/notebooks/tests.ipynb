```python
import torch

# Create a tensor with requires_grad=True (tells autograd to track it)
x = torch.tensor([2.0], requires_grad=True)
w = torch.tensor([3.0], requires_grad=True)
b = torch.tensor([1.0], requires_grad=True)

# Forward pass (autograd records these operations)
y = w * x + b  # y = 3*2 + 1 = 7

# Compute loss (example)
loss = y ** 2  # loss = 49

# Backward pass (autograd computes gradients automatically)
loss.backward()

# Now gradients are stored in .grad attribute
print(f"x.grad = {x.grad}")      # How much loss changes w.r.t. x
print(f"w.grad = {w.grad}")      # How much loss changes w.r.t. w
print(f"b.grad = {b.grad}")      # How much loss changes w.r.t. b
```
